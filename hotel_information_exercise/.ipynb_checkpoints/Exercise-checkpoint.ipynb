```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import seaborn as sns
```

```python
df = pd.read_excel('hotel.xlsx', index_col=[2,3,8])
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

Removing all columns that are not required

```python
df.drop(labels = "Normalized hotel rate (Price / Average of hotel rates)", axis = 1, inplace = True)
```

```python
df.drop(labels = "Normalized Booking Rate (NBR)", axis = 1, inplace = True)
```

```python
df
```

```python
Bookings_more_than_10 = df[df['Count of room type'] >= 10]
```

```python
Average_room_rate = Bookings_more_than_10.groupby(["city_id", "hotel_id", "Room type"])['ADR_USD'].mean()
Average_room_rate_df = pd.DataFrame(Average_room_rate)
```

```python
Average_room_rate_df
```

```python
Bookings_more_than_10["Average_room_rate"] = Average_room_rate_df["ADR_USD"]
```

```python
Bookings_more_than_10["Normalized hotel rate (Price / Average of hotel room rates)"] =Bookings_more_than_10["ADR_USD"]  / Bookings_more_than_10["Average_room_rate"] 
```

```python
Bookings_more_than_10
```

```python
np.quantile(df['Pre-booked Duration'],[0.25])
```

```python
Bookings_more_than_10.plot.scatter(x = "Pre-booked Duration", y = "Normalized hotel rate (Price / Average of hotel room rates)")
Bookings_more_than_10.plot.scatter(x = "Period Popularity", y = "Normalized hotel rate (Price / Average of hotel room rates)")
```

```python
Bookings_more_than_10.boxplot(column = "Normalized hotel rate (Price / Average of hotel room rates)", by =  ["Period Popularity"])
Bookings_more_than_10.boxplot(column = "Normalized hotel rate (Price / Average of hotel room rates)", by =  ["Pre-booked Duration"])
```

```python
Bookings_more_than_10.groupby('star_rating').boxplot(column = "Normalized hotel rate (Price / Average of hotel room rates)"
                                                     , by =  ["Period Popularity.1"]
                                                     , figsize = [12,10]
                                                     , rot = 45)
```

```python
Bookings_more_than_10.groupby('chain_hotel').boxplot(column = "Normalized hotel rate (Price / Average of hotel room rates)"
                                                     , by =  ["Period Popularity"]
                                                     , figsize = [12,10]
                                                     , rot = 45)
```
